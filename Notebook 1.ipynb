```
import math
import random

def linear(net):
    return net

def reLU(net):
    return max(0, net)

def sigmoid(net):
    return 1 / (1 + math.exp((-1) * net))
```

```
f = open("model3.txt", "r")

activation = f.readline().strip().split(',')
neuron = f.readline().strip().split(',')
neuron = [int(x) for x in neuron]
weight = []

for i in range(len(neuron) - 1):
    weight.append([])
    for j in range(neuron[i] + 1):
        data = f.readline().strip().split(',')
        data = [int(x) for x in data]
        weight[i].append(data)

f.close()
print(weight)
print(activation, neuron)

inp = []
num = input("Jumlah instances:")
for i in range(int(num)):
    vars = []
    for j in range(len(weight[0])-1):
        var = input("Instances " + str(i+1) + " Variable " + str(j+1) + ": ")
        vars.append(var)
    inp.append(vars)
print(inp)
```

```
# h = [[]]
# # for i in range(len(inp[0]) - 1):
# #     h[0].append(random.randint(0,10))

# h[0].append(1)
# h[0].append(1)
# # print(h)

inp = [[0, 0], [0, 1]]

# print(inp)
for item in inp:
    h = [item]
    for i in range(len(weight)):
        h.append([0 for j in range(neuron[i + 1])])
        net = [0 for j in range(neuron[i + 1])]
        for k in range(len(weight[i][0])):
            net[k] += weight[i][0][k]
        for j in range(1, len(weight[i])):
            for k in range(len(weight[i][j])):
                net[k] += h[i][j - 1] * weight[i][j][k]
                
        for j in range(len(net)):
            h[i + 1][j] = eval(activation[0] + "({})".format(net[j]))

    print(h[0], end=" ")
    if (h[len(h)-1][0]) > 0.5:
        print("kelas positif")
    else:
        print("kelas negatif")
```

```
hiddenLayer = len(neuron)-2
print("x("+ str(neuron[0]) +")", end=" --> ")
for i in range(hiddenLayer):
    print("h"+str(i+1)+"("+ str(neuron[i+1]) +")", end=" --> ")
print("y("+ str(neuron[len(neuron)-1]) +")\n")

print("Weight:")
print("Input layer:")
if (hiddenLayer == 0):
    for i in range(int(neuron[1])):
        print("b -> y"+str(i+1)+" = "+ str(weight[0][0][i]))
    for i in range(int(neuron[0])):
        for j in range(int(neuron[1])):
            print("x"+ str(i+1) +" -> y"+ str(j+1)+" = "+ str(weight[0][i+1][j]))
else:
    for i in range(int(neuron[1])):
        print("b -> h1"+str(i+1)+" = "+ str(weight[0][0][i]))
    for i in range(int(neuron[0])):
        for j in range(int(neuron[1])):
            print("x"+ str(i+1) +" -> h1"+ str(j+1)+" = "+ str(weight[0][i+1][j]))

if hiddenLayer != 0:
    print("\nHidden Layer:")
    for i in range(hiddenLayer):
        if (i+1 == hiddenLayer):
            print("h"+str(i+1))
            for m in range(int(neuron[i+2])):
                print("b -> y"+ str(m+1) + " = "+ str(weight[i+1][0][m]))
            for n in range(int(neuron[i+1])):
                for k in range(int(neuron[i+2])):
                    print("h"+ str(i+1) + str(n+1) +" -> y"+ str(k+1) +" = "+ str(weight[i+1][n+1][k]))
        else:
            print("h"+str(i+1))
            for m in range(int(neuron[i+2])):
                print("b -> h"+ str(i+2) + str(m+1)+" = "+ str(weight[i+1][0][m]))
            for n in range(int(neuron[i+1])):
                for k in range(int(neuron[i+2])):
                    print("h"+ str(i+1) + str(n+1) +" -> h"+ str(i+2)+str(k+1)+" = "+ str(inp[i+1][n+1][k]))

print("\nActivation function:")
for i in range(len(activation)):
    if i == 0:
        if len(activation) == 1:
            print("x --> y = " + activation[i])
        else:
            print("x --> h1 = " + activation[i])
    elif i == len(activation)-1:
        print("h"+ str(i) +" --> y = " + activation[i])
    else:
        print("h"+ str(i) +" --> h"+ str(i+1) + " = " + activation[i])
```
